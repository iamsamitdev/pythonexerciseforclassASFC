```python
plt.style.use('seaborn')

y = [10, 15, 15, 20, 10]
x = np.arange(1, 6)

bw = 0.25  #

# 
plt.bar(x, y, color='g', width=bw, label=' #1')

#  X  ()
y = [15, 25, 10, 15, 20]
plt.bar(x + bw, y, color='lightblue', width=bw, label=' #2')

#  X 
y = [10, 20, 15, 10, 15]
plt.bar(x + (2 * bw), y, hatch='..', color='lightgray', 
        width=bw, label=' #3')

# tick  X 
plt.xticks(x + bw, [2016, 2017, 2018, 2019, 2020])

plt.rcParams['font.sans-serif'] = 'tahoma'
plt.rcParams['font.family'] = 'tahoma'
plt.rcParams['font.size'] = 14
plt.legend(loc='best')

plt.show()
```

```python
plt.style.use('seaborn')
plt.rcParams['font.sans-serif'] = 'tahoma'
plt.rcParams['font.family'] = 'tahoma'
plt.rcParams['font.size'] = 14

data = {' 1': [10, 20, 40, 30, 30],
        ' 2': [40, 20, 5, 25, 15],
        ' 3': [25, 30, 25, 40, 20] }

idx = ['', '', '', '', '']
df = pd.DataFrame(data, index=idx)

df.plot(kind='line', 
        style=['ro-', 'gD:', 'bs--'],
        use_index=True, legend=True, 
        rot=45, fontsize=14)

plt.title(' - ', fontsize=16, c='g')
plt.xlabel('', fontsize=16, c='m')
plt.ylabel('', fontsize=16, c='m')
```

```python
plt.style.use('seaborn')
plt.rcParams['font.sans-serif'] = 'tahoma'
plt.rcParams['font.family'] = 'tahoma'
plt.rcParams['font.size'] = 14

data = {' 1': [10, 20, 40, 30, 30],
        ' 2': [40, 20, 5, 25, 15],
        ' 3': [25, 30, 25, 40, 20]}

idx = ['', '', '', '', '']

df = pd.DataFrame(data, index=idx)

df.plot(kind='bar', rot=0, fontsize=14, legend=True, 
        color=['green', 'lightblue', 'blue'])

plt.title(' - ', fontsize=16, c='g')
plt.xlabel('', fontsize=16, c='m')
plt.ylabel('', fontsize=16, c='m')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = {'one': np.random.randint(10, 101, 5),
        'two': np.random.randint(10, 101, 5),
        'three': np.random.randint(10, 101, 5)}

df = pd.DataFrame(data, index=list('ABCDE'))

exp = [0, 0.2, 0, 0, 0.2]
df['two'].plot.pie(fontsize=14, autopct='%.1f%%', 
                   startangle=45, shadow=False,
                   explode=exp, 
                   colors=['lightgreen', 'lightcyan', 'pink', 
                           'lightblue', 'wheat'])

plt.ylabel('')
```
